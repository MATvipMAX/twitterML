```python
import numpy as np
import pandas as pd
from utils.users import load_users, select_post, combine_users
from utils.tokeniser import Tokenizer, token_text
from nltk.tokenize import TweetTokenizer
import matplotlib.pyplot as plt
import seaborn as sns


DIR = '../data/'

POST_SIZE = 100
MAX_POSTS = 3200
NB_DEPRESS = 1984

inputs, labels = load_users(DIR, POST_SIZE, MAX_POSTS, NB_DEPRESS, True)
inputs = inputs.drop_duplicates()
control = inputs[inputs.label == 0].reset_index(drop=True)
depress = inputs[inputs.label == 1].reset_index(drop=True)
```

```python
tknzr = TweetTokenizer(reduce_len=True)
inputs['token'] = inputs.clean.apply(lambda x: tknzr.tokenize(x.lower()))
inputs['len'] = inputs.token.apply(lambda x: len(x))
```

```python
group_names =['4.3M posts', '5.5M posts']
group_size =[con_post_nb, dep_post_nb]
subgroup_names =[str(con_user_nb) + ' users', str(dep_user_nb) + ' users']
subgroup_size =[con_user_nb, dep_user_nb]

a, b, c=[plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(16, 7))
plt.rcParams.update({'font.size': 12})

ax1.set_title('The Ratio of Retweets to Tweet', pad=79.0)
# ax1.text(-.90, 1.86, 'The Ratio of Retweets to Tweet', fontsize=14)

# ax.axis('equal')
mypie, _ = ax1.pie(group_size, radius=1.5, labels=group_names, labeldistance=0.75, colors=[a(0.6), b(0.6), c(0.6)] )
plt.setp( mypie, width=0.49, edgecolor='white')

mypie2, _ = ax1.pie(subgroup_size, radius=1.3-0.3, labels=subgroup_names, labeldistance=0.6, colors=[a(0.4), b(0.4)])
plt.setp( mypie2, width=0.5, edgecolor='white')
# plt.margins(0,0)

#################################### Plot The Ratio of Retweets to Tweet
data = [control.groupby(['userid']).sum().reset_index().rt.values/control.groupby(['userid']).count().reset_index().rt.values,
       depress.groupby(['userid']).sum().reset_index().rt.values/depress.groupby(['userid']).count().reset_index().rt.values]

ax2.set_title('The Ratio of Retweets to Tweet')
box = ax2.boxplot(data, showfliers=False, patch_artist=True, widths=0.5)

colors = [a(0.6), b(0.6)]
 
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    
for element in ['whiskers', 'fliers', 'means', 'medians', 'caps']:
    plt.setp(box[element], color='black')
    
ax2.yaxis.grid(True)
ax2.set_xlabel('Classes')
ax2.set_ylabel('Ratio')

#################################### Plot The Number of Words per Tweet
data = [control.len.values,
       depress.len.values]

ax3.set_title('The Number of Words per Tweets')
box = ax3.boxplot(data, showfliers=False, patch_artist=True, widths=0.5)

colors = [a(0.6), b(0.6)]
 
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    
for element in ['whiskers', 'fliers', 'means', 'medians', 'caps']:
    plt.setp(box[element], color='black')
    
ax3.yaxis.grid(True)
ax3.set_xlabel('Classes')
ax3.set_ylabel('Words')



plt.subplots_adjust(wspace=.4)
plt.show()
```

```python

# import matplotlib as mpl
mpl.style.use('default')
plt.style.use('default')

nice_fonts = {
        # Use LaTeX to write all text
        "font.family": "serif",
        # Use 10pt font in plots, to match 10pt font in document
        "axes.labelsize": 10,
        "font.size": 10,
        # Make the legend/label fonts a little smaller
        "legend.fontsize": 10,
        "xtick.labelsize": 10,
        "ytick.labelsize": 10,
}

plt.rcParams.update(nice_fonts)

f, axes = plt.subplots(1, 3, figsize=(15, 5))



def func(pct, allvals):
    absolute = float(pct/1000.*np.sum(allvals))
    return "{:.1f} M".format(absolute)


def func2(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d} users".format(absolute)


size = 0.3
vals = np.array([[43, 55]])
user_nb = np.array([con_user_nb, dep_user_nb])

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.array([2, 6, 9, 10]))

axes[0].pie(vals, radius=1, colors=outer_colors, autopct=lambda pct: func(pct, vals), pctdistance=.85,
       wedgeprops=dict(width=size, edgecolor='w'))

axes[0].pie(user_nb, radius=1-size, colors=inner_colors, autopct=lambda pct: func2(pct, user_nb), pctdistance=.75,
       wedgeprops=dict(width=size, edgecolor='w'))

axes[0].set(aspect="equal")
axes[0].set_title('The Number of Users and Tweets', y=1.06)

c = control.groupby(['userid']).sum().reset_index().rt.values/control.groupby(['userid']).count().reset_index().rt.values
c = pd.DataFrame(c, columns=['Retweet'])
c['Classes'] = 'Control'

d = depress.groupby(['userid']).sum().reset_index().rt.values/depress.groupby(['userid']).count().reset_index().rt.values
d = pd.DataFrame(d, columns=['Retweet'])
d['Classes'] = 'Depression'

comb = pd.concat([c, d]).reset_index(drop=True)

sns.boxplot(x="Classes", y="Retweet", data=comb, showfliers=False, ax=axes[2], width=0.5).set_title("The Ratio of Retweets to Tweets")
axes[2].set_ylabel('Ratio')

c = control.len.values
c = pd.DataFrame(c, columns=['Words'])
c['Classes'] = 'Control'

d = depress.len.values
d = pd.DataFrame(d, columns=['Words'])
d['Classes'] = 'Depression'

comb = pd.concat([c, d]).reset_index(drop=True)

sns.boxplot(x="Classes", y="Words", data=comb, showfliers=False, ax=axes[1], width=0.5).set_title("The Number of Words per Tweet")

f.savefig('data_des', dpi=300)
f.show()
```

```python

f, axes = plt.subplots(1, 3, figsize=(15, 5))


def func(pct, allvals):
    absolute = float(pct/1000.*np.sum(allvals))
    return "{:.1f} M".format(absolute)


def func2(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}".format(absolute)


size = 0.3
vals = np.array([[41, 49]])
user_nb = np.array([1755, 2001])

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.array([2, 6, 9, 10]))

axes[0].pie(vals, radius=1, colors=outer_colors, autopct=lambda pct: func(pct, vals), pctdistance=.85,
       wedgeprops=dict(width=size, edgecolor='w'))

axes[0].pie(user_nb, radius=1-size, colors=inner_colors, autopct=lambda pct: func2(pct, user_nb), pctdistance=.75,
       wedgeprops=dict(width=size, edgecolor='w'))

axes[0].set(aspect="equal")
axes[0].set_title('The Number of Users and Tweets', y=1.05)

c = control.groupby('userid').mean().retweet.values
c = pd.DataFrame(c, columns=['Retweet'])
c['Class'] = 'Control'

d = depress.groupby('userid').mean().retweet.values
d = pd.DataFrame(d, columns=['Retweet'])
d['Class'] = 'Depression'

comb = pd.concat([c, d]).reset_index(drop=True)

sns.boxplot(x="Class", y="Retweet", data=comb, showfliers=False, ax=axes[2]).set_title("The Ratio of Retweet to Tweets")
axes[2].set_ylabel('Ratio')

c = control.groupby('userid').mean().words.values
c = pd.DataFrame(c, columns=['Words'])
c['Class'] = 'Control'

d = depress.groupby('userid').mean().words.values
d = pd.DataFrame(d, columns=['Words'])
d['Class'] = 'Depression'

comb = pd.concat([c, d]).reset_index(drop=True)

sns.boxplot(x="Class", y="Words", data=comb, showfliers=False, ax=axes[1]).set_title("The Number of Words")

f.savefig('data_des', dpi='figure')
```
